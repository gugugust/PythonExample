```
# 과제 예제 실행
import re
txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt)
print(x)
```

```
# 과제 예제 실행
import re
txt = "The rain in Spain"
x = re.findall("ai", txt)
print(x)
```

```
# 과제 예제 실행
import re
txt = "The rain in Spain"
x = re.search("\s", txt)
print("The first white-space character is located in position:", x.start())
```

```
# 과제 예제 실행
import re
txt = "The rain in Spain"
x = re.split("\s", txt)
print(x)
```
